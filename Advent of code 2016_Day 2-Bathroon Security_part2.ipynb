```python


class AbstractKey:
    def __init__(self, num):
        self.num = num
    def right(self):
        pass
    def left(self):
        pass
    def up(self):
        pass
    def down(self):
        pass

    def print(self):
        print(self.num)

class Key(AbstractKey):

    def __init__(self, num):
        super().__init__(num)
    
    def right(self):
        return lookup_key[self.num + 1]() # Will return Key()

    def left(self):
        return lookup_key[self.num - 1]()

    def up(self):
        return lookup_key[self.num - 4]()

    def down(self):
        return lookup_key[self.num + 4]()

class UpperKey(Key):
    def __init__(self, num):
        super().__init__(num)

    def up(self):
        return self # self returns This key
    
class LowerKey(Key):
    def __init__(self, num):
        super().__init__(num)
    
    def down(self):
        return self

class Key1(UpperKey):
    def __init__(self):
        super().__init__(1)
    
    def left(self):
        return self
    def right(self):
        return self
    def down(self):
        return lookup_key[self.num + 2]() 

class Key2(UpperKey):
    def __init__(self):
        super().__init__(2)
    def left(self):
        return self

class Key3(Key):
    def __init__(self):
        super().__init__(3)
    def up(self):
        return lookup_key[self.num - 2]() # 
    

class Key4(UpperKey):
    def __init__(self):
        super().__init__(4)
    
    def right(self):
        return self

class Key5(UpperKey):
    def __init__(self):
        super().__init__(5)
    def left(self):
        return self
    def down(self):
        return self

class Key6(Key):
    def __init__(self):
        super().__init__(6)
    def down(self):
        return lookup_key['A']()

class Key7(Key):
    def __init__(self):
        super().__init__(7)
    
    def down(self):
        return lookup_key['B']()

class Key8(LowerKey):
    def __init__(self):
        super().__init__(8)
    def down(self):
        return lookup_key['C']()

class Key9(LowerKey):
    def __init__(self):
        super().__init__(9)
    def right(self):
        return self
    def up(self):
        return self

class KeyA (LowerKey):
    def __init__(self):
        super().__init__('A')
    def left (self):
        return self
    def up(self):
        return lookup_key[6]()
    def right(self):
        return lookup_key['B']()
    

class KeyB (Key):
    def __init__(self):
        super().__init__('B')
    def left (self):
        return lookup_key['A']()
    def right (self):
        return lookup_key['C']()
    def up (self):
        return lookup_key[7]()
    def down (self):
        return lookup_key['D']()

class KeyC(LowerKey):
    def __init__(self):
        super().__init__('C')
    def up(self):
        return lookup_key[8]()
    def right(self):
        return self
    def left(self):
        return lookup_key['B']()
    
    
class KeyD(LowerKey):
    def __init__(self):
        super().__init__('D')
    def up(self):
        return lookup_key['B']()
    def left (self):
        return self
    def right (self):
        return self
    


lookup_key = {1: Key1, 2: Key2, 3 : Key3, 4: Key4, 5: Key5, 6: Key6, 7: Key7, 8: Key8, 9: Key9, 'A': KeyA, 'B': KeyB, 'C':KeyC, 'D':KeyD}

input_string= ""
current_key = Key5()
for c in input_string:
    if c == 'U':
        current_key = current_key.up()
    elif c == 'D':
        current_key = current_key.down()
    elif c == 'L':
        current_key = current_key.left()
    else:
        current_key = current_key.right()
current_key.print()

```

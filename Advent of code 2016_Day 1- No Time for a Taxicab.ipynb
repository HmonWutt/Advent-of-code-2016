```python
class Santa:
    def __init__(self):
        self.current_x = 0
        self.current_y = 0
        self.direction = 'N'
        self.path = []

    def update_direction(self, turn):
        if self.direction == "N":
            if turn == 'R':
                self.direction = 'E'
            else:
                self.direction = 'W'

        elif self.direction == 'S':
            if turn == 'R':
                self.direction = 'W'
            else:
                self.direction = 'E'

        elif self.direction == 'E':
            if turn == 'R':
                self.direction = 'S'
            else:
                self.direction = 'N'
        else:
            if turn =='R':
                self.direction = 'N'
            else:
                self.direction = 'S'
        
        return self.direction

    def update_position (self, coordinate):
        coordinate = int(coordinate)
        if self.direction == 'E':
            self.step_x_plus(coordinate)
        elif self.direction == 'W':
            self.step_x_minus(coordinate)
        elif self.direction == 'N':
            self.step_y_plus(coordinate)
        else:
            self.step_y_minus(coordinate)
        return self.current_x, self.current_y

    def step_x_plus(self, coordinate):
        for _ in range(coordinate):
            self.current_x += 1
            if (self.current_x, self.current_y) in self.path:
                print("Found place")
                self.print_position()
                self.print_abs_pos()
            else:
                self.path.append((self.current_x, self.current_y))

    def step_x_minus(self, coordinate):
        for _ in range(coordinate):
            self.current_x -= 1
            if (self.current_x, self.current_y) in self.path:
                print("Found place")
                self.print_position()
                self.print_abs_pos()
            self.path.append((self.current_x, self.current_y))

    def step_y_plus(self, coordinate):
        for _ in range(coordinate):
            self.current_y += 1
            if (self.current_x, self.current_y) in self.path:
                print("Found place")
                self.print_position()
                self.print_abs_pos()
            self.path.append((self.current_x, self.current_y))
    
    def step_y_minus(self, coordinate):
        for _ in range(coordinate):
            self.current_y -= 1
            if (self.current_x, self.current_y) in self.path:
                print("Found place")
                self.print_position()
                self.print_abs_pos()
            self.path.append((self.current_x, self.current_y))
            

    def print_position(self):
        print(self.current_x, self.current_y)
    
    def print_abs_pos(self):
        print(abs(self.current_x) + abs(self.current_y))

```

```python
santa = Santa()
santa.update_direction('R')
santa.update_position(5)
santa.print_position()
santa.print_abs_pos()
```
